```python
class BaseModels():
    def __init__(self, X_train, y_train, X_test, y_test):
        self.scaler = StandardScaler()

        self.X_train = self.scaler.fit_transform(X_train).astype(np.float32)
        self.X_test = self.scaler.transform(X_test).astype(np.float32)

        # Convert y to numpy array and float32
        self.y_train = np.asarray(y_train).astype(np.float32)
        self.y_test = np.asarray(y_test).astype(np.float32)

        # XGBoost with custom objective
        def xgb_custom_obj(y_pred, y_true):
            grad = y_pred - y_true
            hess = np.ones_like(y_pred)
            return grad, hess

        self.xgb_model = XGBRegressor(
            device='cuda',
            objective=xgb_custom_obj,
            n_estimators=1000,
            learning_rate=0.001,
            max_depth=6,
            min_child_weight=1,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42
        )
        class CatBoostCustomLoss:
            def calc_ders_range(self, approxes, targets, weights):
                grad = approxes - targets
                hess = np.ones_like(approxes)
                return grad, hess

        def custom_metric(approxes, target, weight):
            return np.mean(np.abs(approxes - target))

        self.cat_model = CatBoostRegressor(
            task_type='GPU',
            loss_function='RMSE',
            eval_metric='MAE',
            iterations=1000,
            learning_rate=0.05,
            depth=6,
            l2_leaf_reg=3,
            random_seed=42,
            verbose=100
        )


        # LightGBM with custom objective
        # def lgb_custom_obj(y_pred, y_true):
        #     grad = y_pred - y_true
        #     hess = np.ones_like(y_pred)
        #     return grad, hess
        # self.lgb_model = LGBMRegressor(
        #             #device = 'gpu',
        #             # gpu_platform_id = 0,
        #             # gpu_device_id = 0,
        #             objective=lgb_custom_obj,
        #             n_estimators=1000,
        #             learning_rate=0.01,
        #             num_leaves=31,
        #             subsample=0.8,
        #             colsample_bytree=0.8,
        #             random_state=42
        # )

        # SVR (instead of SVC, as this is a regression problem)
        self.svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)

        # Meta-learner (Stacking Regressor)
        self.meta_learner = StackingRegressor(
            estimators=[
                ('xgb', self.xgb_model),
                ('cat', self.cat_model),
                # ('lgb', self.lgb_model),
                ('svr', self.svr_model)
            ],
            final_estimator=Ridge(alpha=1.0),
            cv=5
        )

    def fit_xgb(self):
        self.xgb_model.fit(self.X_train, self.y_train)
        self._evaluate_model(self.xgb_model, "XGBoost")

    def fit_cat(self):
        self.cat_model.fit(
            self.X_train, self.y_train,
            eval_set=(self.X_test, self.y_test),
            use_best_model=True,
            early_stopping_rounds=50
        )
        self._evaluate_model(self.cat_model, "CatBoost")

    def fit_lgb(self):
        self.lgb_model.fit(self.X_train, self.y_train)
        self._evaluate_model(self.lgb_model, "LightGBM")

    def fit_svr(self):
        self.svr_model.fit(self.X_train, self.y_train)
        self._evaluate_model(self.svr_model, "SVR")

    def fit_all(self):
        # self.fit_lgb()
        self.fit_xgb()
        self.fit_cat()
        self.fit_svr()

    def fit_meta_learner(self):
        self.meta_learner.fit(self.X_train, self.y_train)
        self._evaluate_model(self.meta_learner, "Meta-learner")

    def _evaluate_model(self, model, name):
        y_pred = model.predict(self.X_test)
        mse = mean_squared_error(self.y_test, y_pred)
        mae = mean_absolute_error(self.y_test, y_pred)
        r2 = r2_score(self.y_test, y_pred)
        cv_score = np.mean(cross_val_score(model, self.X_train, self.y_train, cv=5, scoring='neg_mean_squared_error'))

        print(f"\n{name} Evaluation:")
        print(f"Mean Squared Error: {mse}")
        print(f"Mean Absolute Error: {mae}")
        print(f"R2 Score: {r2}")
        print(f"Cross-validation MSE: {-cv_score}")

    def predict(self):
        xgb_preds = self.xgb_model.predict(self.X_test)
        cat_preds = self.cat_model.predict(self.X_test)
        #lgb_preds = self.lgb_model.predict(self.X_test)
        svr_preds = self.svr_model.predict(self.X_test)
        meta_preds = self.meta_learner.predict(self.X_test)

        return xgb_preds, cat_preds, lgb_preds, svr_preds, meta_preds

# Usage
base_models = BaseModels(X, Y, X_test, Y_test)
base_models.fit_all()
base_models.fit_meta_learner()
xgb_preds, cat_preds, lgb_preds, svr_preds, meta_preds = base_models.predict()
```

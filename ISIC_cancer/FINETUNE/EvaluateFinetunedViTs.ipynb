```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/Users/jimmyhe/Desktop/KaggleCompetitions/ISISCANCER/models/finetuned_model_facebook_deit-base-distilled-patch16-224/loss_log_facebook_deit-base-distilled-patch16-224.csv')

# Plot Loss
plt.figure(figsize=(10, 5))
plt.plot(df1['epoch'], df1['train_loss'], label='Train Loss')
plt.plot(df1['epoch'], df1['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('facebook_deit-base-distilled-patch16-224 Training and Validation Loss')
plt.legend()
plt.show()

# Plot Accuracy
plt.figure(figsize=(10, 5))
plt.plot(df1['epoch'], df1['train_accuracy'], label='Train Accuracy')
plt.plot(df1['epoch'], df1['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()

# Plot PAUC
plt.figure(figsize=(10, 5))
plt.plot(df1['epoch'], df1['train_pauc'], label='Train PAUC')
plt.plot(df1['epoch'], df1['val_pauc'], label='Validation PAUC')
plt.xlabel('Epoch')
plt.ylabel('PAUC')
plt.title('Training and Validation PAUC')
plt.legend()
plt.show()

# Plot MSE
plt.figure(figsize=(10, 5))
plt.plot(df1['epoch'], df1['train_MSE'], label='Train MSE')
plt.plot(df1['epoch'], df1['val_MSE'], label='Validation MSE')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.title('Training and Validation MSE')
plt.legend()
plt.show()
```

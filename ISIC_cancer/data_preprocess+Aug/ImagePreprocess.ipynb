```python
df_train_metadata = pd.read_csv('/Users/jimmyhe/Desktop/KaggleCompetitions/ISISCANCER/isic-2024-challenge/train-metadata.csv')
df_test_metadata = pd.read_csv('/Users/jimmyhe/Desktop/KaggleCompetitions/ISISCANCER/isic-2024-challenge/test-metadata.csv')
path_list = [f"/Users/jimmyhe/Desktop/KaggleCompetitions/ISISCANCER/isic-2024-challenge/train-image/image/{id}.jpg" for id in df_train_metadata.isic_id]
path_list = path_list[:10]
IDs = df_train_metadata.isic_id.values
IDs = IDs[:10]
```

```python
if __name__ == "__main__":
    output_dir = "/Users/jimmyhe/Desktop/KaggleCompetitions/ISISCANCER/preprocessed_image_peek"
    os.makedirs(output_dir, exist_ok=True)

    args_list = [(path_list[i], id, output_dir) for i, id in enumerate(IDs)]

    num_processes = os.cpu_count()

    with Pool(processes=num_processes) as pool:
        list(tqdm(pool.imap(process_image, args_list), total=len(args_list), desc="Processing images"))
```

```python
def display_images_path(path_list,num_to_display=10):
  for i in range(num_to_display):
    image_path = path_list[i]
    img = mpimg.imread(image_path)
    plt.figure()
    plt.imshow(img)
    plt.title(image_path)
    plt.show()
```
